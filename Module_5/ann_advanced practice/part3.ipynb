```python
import time
weightToTry = np.linspace(-5,5,1000)
costs = np.zeros((1000,1000))

startTime = time.clock()
for i in range(1000):
    for j in range(1000):
        NN.W1[0,0] = weightToTry[i]
        NN.W1[0,1] = weightToTry[j]
        yHat = NN.forwardPropagation(X)
        costs[i,j] = 0.5*sum((y-yHat)**2)
endTime = time.clock()
```
